cleaning and filling the missing values in the data

```python
data['Caffeine (mg)'] = data['Caffeine (mg)'].replace('Varies', np.NaN).replace('varies', np.NaN)
data['Caffeine (mg)'] = data['Caffeine (mg)'].astype(np.float64)
data['Caffeine (mg)'] = data['Caffeine (mg)'].fillna(data['Caffeine (mg)'].mean())
```

```python
data['Total Fat (g)'].unique()
```

```python
data['Total Fat (g)'] = data['Total Fat (g)'].replace('3 2', '3.2')
```

```python
data.info()
```

```python
# Extract columns with int and float types
numeric_columns = data.select_dtypes(include=["int64", "float64"]).columns
data_numeric = data[numeric_columns]
```

We will be analysing the dataset using the fact that weather the drink comes under the category tea or not

```python
data['Beverage_category'].unique()
```

```python
data['Tea'] = data['Beverage_category'].apply(lambda x: 1 if x == 'Tazo Tea Drinks' else 0)
data = data.drop('Beverage_category', axis=1)
```

```python

# Ensure 'Tea' column is of numeric type (int or float)
data['Tea'] = data['Tea'].astype(int)
```

```python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data = onehot_encode(
    data,
    columns=['Beverage', 'Beverage_prep'],
    prefixes=['bev', 'bevp']
)
```

```python

data = data.replace({True: 1, False: 0})
```

```python
data = data.applymap(lambda x: np.float64(str(x).replace('%', '')))
```

```python
## to do
# Create a pie chart of the 'Tea' column also write your observation form the plot

# Calculate the counts of different tea preferences
counts = data['Tea'].value_counts()
# Plotting a pie chart
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=['Non-Tea', 'Tea'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Tea vs Non-Tea Drinks')
plt.show()
```

**Observation from the Plot:**


The pie chart describes the proportion of drinks

```python
## to do
# perform pca on the data and plot the explained variace ratio, what is the optimal number of principal components in this case ?

# Select only numeric columns for PCA
columns = data.select_dtypes(include=['int64', 'float64']).columns
X = data[columns]
# Ensure all numeric columns are of the correct dtype
X = X.apply(pd.to_numeric, errors='coerce')

# Handle any remaining NaN values
X.fillna(0, inplace=True)  # Replace NaN with 0, adjust as needed for your dataset

# Now X should contain only numeric columns suitable for PCA
# Perform PCA
pca = PCA()
pca.fit(X)

# Plotting the explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance Ratio by Principal Components')
plt.grid(True)
plt.show()
```

**Observation from the plot:**

The plot of cumulative explained variance will help us determine the optimal number of principal components to retain based on the variance explained.

```python
## to do
# visualise the principal components, choose the number of principal components based on the above plot. What is you observation from the plot?

#Based on the previous plot, choose an appropriate number of principal components and visualize them:
# Example: Visualizing the first two principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.8)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Principal Components 1 vs 2')
plt.grid(True)
plt.show()
```

**Observation from the Plot:**

*   Analyze how the data points are
distributed in the reduced-dimensional space.
*  Observe any clusters or patterns that emerge among the drinks.

```python
## to do
# plot the first 2 components of tsne, whats you observation from the plot?

from sklearn.manifold import TSNE
# Perform t-SNE
tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X)
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], alpha=0.8)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE Components 1 vs 2')
plt.grid(True)
plt.show()
```

**Observation from the Plot:**


*   Evaluate any clustering or separation patterns that may differ from PCA results.
*   Identify any distinct groupings among the Starbucks drinks.
